```python

# Definición de la función
def f(x):
    return 1 + (1 / 2) * mt.tanh(2 * x)

# Función de diferencia central
def central(x, h, f):
    return (f(x + h) - f(x - h)) / (2 * h)

# Rango de x
valores_x = np.linspace(-2, 2, 100)
h = 0.01

# Calcular derivada numérica
derivada_numerica = [central(x, h, f) for x in valores_x]

# Definición de la derivada analítica
def derivada_analitica(x):
    return 1 / mt.cosh(2 * x)**2

# Calcular derivada analítica para todos los valores de x
valores_derivada_analitica = [derivada_analitica(x) for x in valores_x]

# Graficar
plt.figure(figsize=(10, 6))
plt.plot(valores_x, derivada_numerica, 'o', label='Derivada Numérica', markersize=4)
plt.plot(valores_x, valores_derivada_analitica, '-', label='Derivada Analítica', linewidth=2)
plt.title('Comparación de Derivadas Numérica y Analítica')
plt.xlabel('x')
plt.ylabel("f'(x)")
plt.legend()
plt.grid()
plt.show()
```
